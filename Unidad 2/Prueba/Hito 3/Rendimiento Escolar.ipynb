## Visualización de Conjunto de datos
Partiremos por refactorizar los datos conseguidos

```python
df = pd.read_csv('students.csv', delimiter='|').drop(columns=['Unnamed: 0'])
parsedDf = df.copy()
# Parseamos todas las columnas:
for col in df.columns:
    if parsedDf[col].dtype == 'object':
        # Se quitan las comillas dobles y se transforman los datos nulos
        parsedDf[col] = parsedDf[col].apply(lambda x: x.replace('"', '')).replace(['nulidade', 'sem validade', 'zero'], [float("NaN"), float("NaN"),float("NaN")])
# Convertimos aquellas variables 
intCols = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'absences', 'G1', 'G3']
for intCol in intCols:
    parsedDf[intCol] = pd.to_numeric(parsedDf[intCol], downcast='integer')
parsedDf.dtypes
```

Hacemos un Análisis de los datos perdidos

```python
# Vemos las variables perdidas
msngo.matrix(parsedDf, labels=True)
msngo.matrix(parsedDf.dropna(), labels=True)
# Sólo dejamos los registros completos
parsedDf = parsedDf.dropna()
```

Vemos que hay varias columnas que presentan datos perdidos, siendo 30 la columna con mas pérdida. Luego de quitar todos los registors quedó con 284 observaciones, es decir, 111 perdidas

Ahora que traspásamos aquellas columnas que podían ser números a los tipos correspondientes podemos hacer un análisis de como se ve el vector objetivo según las demas variables.

```python
# Declaramos un nuevo df para que tenga las columnas nuevas
refactoredDf = parsedDf.copy()
# Creamos los graficos para ver la relación de G1, G2, y G3 con el resto de las variables
fig, ax = plt.subplots(len(parsedDf.columns) - 3, 3)
# ajustamos el tamaño de la figura
plt.figure(figsize=(20, 6*len(parsedDf.columns)))
# Iteramos sobre las columnas
for i, col in enumerate(parsedDf.columns):
    if parsedDf[col].dtype == 'object':
        frequency = parsedDf[col].value_counts()
        # Se crean los dummies y se añaden al nuevo Dataframe de refactor
        dummies = pd.get_dummies(parsedDf[col], prefix=col)
        if len(frequency) > 2:
            refactoredDf = pd.concat([refactoredDf, dummies], axis=1)
        else:
            refactoredDf = pd.concat([refactoredDf, dummies.iloc[:,-1]], axis=1)
        # Se grafica la frecuencia de las notas diferenciados según las columnas categóricas
        sns.countplot(x='G1', data=parsedDf, hue=col, ax=ax[i][0])
        sns.countplot(x='G2', data=parsedDf, hue=col, ax=ax[i][1])
        sns.countplot(x='G3', data=parsedDf, hue=col, ax=ax[i][2])
    else:
        # Las variables numéricas las hacemos boxplot
        if col not in ['G2', 'G3', 'G1']:
            sns.boxplot(x='G1', y=col, data=parsedDf, ax=ax[i][0])
            sns.boxplot(x='G2', y=col, data=parsedDf, ax=ax[i][1])
            sns.boxplot(x='G3', y=col, data=parsedDf, ax=ax[i][2])
```

Podemos observar que la gran mayoría de las variables mantienen su distribución entre los tres periodos de notas, y pareciera que normalmente el periodo 'G3' es el que mas diferencia tiene con respecto a 'G2' y 'G1'.

Los análisis mas interesantes son:

<ul>
<li>Para <code>studytime</code> Se puede observar que el primer periodo presenta que casi todos tenían una canntidad de horas altas, pero para los con notas mas bajas este tiempo disminuye hacia el último periodo</li>
<li>Con <code>internet</code> podemos notar que la mayoría de las personas que tienen notas bajo el promedio no tienen acceso a internet.</li>
<li>a traves de <code>Dalc</code> se puede ver que el consumo de alcohol durante la semana aumenta hacia el último periodo en las personas que tienen bajas notas.</li>
</ul>

```python
latDf = parsedDf.loc[:,'famrel':'health']
for i in latDf.columns:
    latDf[i] = latDf[i].astype(int)
factor.calculate_bartlett_sphericity(latDf)
```

```python
pd.Series(factor.calculate_kmo(latDf)[0], index=latDf.columns)
```

Esto nos indica que no es posible extraer valores latentes pues ningún factor presenta un valor superior a 0.7

### Creación de modelos

Analizaremos las correlaciones existentes

```python
plt.figure(figsize=(20, 15))

corrMatrix = refactoredDf.corr()

filteredMatrix = corrMatrix[(corrMatrix > 0.1) | (corrMatrix < -0.1)]

sns.heatmap(filteredMatrix, annot=True)
```

Vemos que para nuestras variables de interés <code>G1, G2 y G3</code> se observa una proporcionalidad directa fuerte entre ellas, seguida por una de bastante menor intensidad <code>Medu</code>. También se aprecia una proporcionalidad inversa considerable con <code>failures</code> seguida por <code>goout</code>.

A continuación haremos las regresiónes lineales para nuestras variables de interés usando todos los factores.

```python
for i in ['G1', 'G2', 'G3']:
    model = smf.ols('{} ~ '.format(i) + '+ '.join(list(filter(lambda x: x != i, parsedDf.columns))), data=parsedDf).fit()
    print('Modelo para {}'.format(i))
    print(model.summary())
```

Podemos observar que los 3 modelos tienen estadísticas de bondad del modelo bastante parecidas. siendo las de <code>G2</code> las mejores.

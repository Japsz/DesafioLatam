# Hito 1, Desafío 2: Rendimiento Escolar


## __Enunciado:__

Lo contactan de una escuela Portuguesa para generar un modelo que identifique aquellos
alumnos que presentan un bajo desempeño académico, medido en el promedio final del año
escolar. Para ello le envían un archivo con registros sociodemográficos y conductuales de
los alumnos de dos escuelas para perfilar a los estudiantes.</br>

De manera adicional la psicopedagoga sugiere inspeccionar una batería de preguntas
asociadas a aspectos ambientales del alumno (de <code>famrel</code> a <code>health</code>) y ver si éstas se
pueden abstraer en categorías latentes.

### __Descripción de la base de datos__
Las variables que componen esta base se detallan a continuación:

<ul>
<li><code>school</code> : Escuela del estudiante. (binaria: 'GP' - Gabriel Pereira o 'MS' - Mousinho da Silveira)</li>

<li><code>sex</code> : Sexo del estudiante. (binaria: 'F' - Mujer o 'M' - Hombre)</li>

<li><code>age</code> : Edad del estudiante. (numérica: de 15 a 22)</li>

<li><code>address</code> : Ubicación de la casa del estudiante. (binaria: 'U' - urbana o 'R' - rural)</li>

<li><code>famsize</code> : Tamaño de la familia. (binaria: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)</li>

<li><code>Pstatus</code> : Estado cohabitacional de los padres. (binaria: 'T' - cohabitando juntos o 'A' - viviendo separados)</li>

<li><code>Medu</code> : Nivel educacional de la madre. (numérica: 0 - ninguno, 1 - educación básica (4to), 2 - de 5to a 9, 3 - educación media, o 4 - educación superior).</li>

<li><code>Fedu</code> : Nivel educacional del padre. (numérica: 0 - ninguno, 1 - educación básica (4to), 2 - de 5to a 9, 3 - educación media, o 4 - educación superior).</li>

<li><code>Mjob</code> : Ocupación de la madre. (nominal: 'teacher' profesora, 'health' relacionada a salud, 'services' (e.g. administración pública o policía), 'at_home' en casa u 'other' otra).</li>

<li><code>Fjob</code> : Ocupación del padre (nominal: 'teacher' profesor, 'health' relacionado a salud, 'services' (e.g. administración pública o policía), 'at_home' en casa u 'other' otra).</li>

<li><code>reason</code> : Razón para escoger la escuela (nominal: 'home' cercano a casa, 'reputation' reputación de la escuela, 'course' preferencia de cursos u 'other' otra)</li>

<li><code>guardian</code> : Apoderado del estudiante (nominal: 'mother' madre, 'father' padre u 'other' otro)</li>

<li><code>traveltime</code> : Tiempo de viaje entre hogar y colegio. (numeric: 1 - <15 min., 2 - 15 a 30 min., 3 - 30 min. a 1 hora, or 4 - >1 hora).</li>

<li><code>studytime</code> : Horas semanales dedicadas al estudio. (numérica: 1 - <2 horas, 2 - 2 a 5 horas, 3 - 5 a 10 horas, o 4 - >10 horas)</li>

<li><code>failures</code> : Número de clases reprobadas. (numérica: n si 1<=n<3, de lo contrario 4)</li>

<li><code>schoolsup</code> : Apoyo educacional del colegio. (binaria: si o no)</li>

<li><code>famsup</code> : Apoyo educacional familiar. (binaria: si o no)</li>

<li><code>paid</code> : Clases particulares pagadas (matemáticas o portugués) (binaria: si o no)</li>

<li><code>activities</code> : Actividades extracurriculares. (binaria: si o no)</li>

<li><code>nursery</code> : Asistió a guardería infantil. (binaria: si o no)</li>

<li><code>higher</code> : Desea proseguir estudios superiores (binaria: si o no)</li>

<li><code>internet</code> : Acceso a internet desde el hogar (binaria: si o no)</li>

<li><code>romantic</code> : Relación romántica (binaria: si o no)</li>

<li><code>famrel</code> : Calidad de las relaciones familiares. (numérica: de 1 - muy malas a 5 - excelentes)</li>

<li><code>freetime</code> : Tiempo libre fuera del colegio (numérica: de 1 - muy poco a 5 - mucho)</li>

<li><code>goout</code> : Salidas con amigos (numérica: de 1 - muy pocas a 5 - muchas)</li>

<li><code>Dalc</code> : Consumo de alcohol en día de semana (numérica: de 1 - muy bajo a 5 - muy alto)</li>

<li><code>Walc</code> : Consumo de alcohol en fines de semana (numérica: de 1 - muy bajo a 5 - muy alto)</li>

<li><code>health</code> : Estado de salud actual (numérica: from 1 - muy malo to 5 - muy bueno)</li>

<li><code>absences</code> : Cantidad de ausencias escolares (numérica: de 0 a 93)</li>

<li><code>G1</code> : Notas durante el primer semestre (numérica: de 0 a 20). Este es uno de sus vectores objetivos para el modelo descriptivo.</li>
<li><code>G2</code> : Notas durante el segundo semestre (numérica: de 0 a 20). Este es uno de sus vectores objetivos para el modelo descriptivo.</li>

<li><code>G3</code> : Promedio final (numérica: de 0 a 20). Este es uno de sus vectores objetivos para el
modelo descriptivo y el vector a predecir en el modelo predictivo.</li>
</ul>

### Aspectos adicionales a considerar
<li>La base de datos presenta una serie de anomalías. En la escuela no tienen buenas
prácticas sobre cómo ingresar datos, por lo que existen datos perdidos que están
registrados bajo tres categorías: nulidade, sem validade, zero. De manera adicional,
hay 3 variables numéricas que se registraron como strings, cuya interpretación en
<code>pandas</code> devuelve una estructura de datos genérica. Finalmente, la base está con un
encoding distinto al normal y los delimitadores son distintos.</li>
<li>Para simplificar el análisis y su posterior inclusión en un modelo predictivo, se
sugiere recodificar las variables binarias como 0 y 1. Se recomienda seguir en
criterio de asignarle 1 a aquellas categorías minoritarias.</li>
<li>El procedimiento también debe aplicarse para aquellas variables nominales con más
de 2 categorías siguiendo la misma lógica.</li>
<li>En la parte de modelación descriptiva, se deben generar modelos utilizando todos
los predictores por cada una de las notas registradas en <code>G1</code>, <code>G2</code> y <code>G3</code>.</li>
<li>Para la parte de modelación predictiva, se debe generar un modelo para predecir las
notas en <code>G3</code>.</li>


## __Preliminares__
 
### __Descripción y Objetivo__:

La idea es perfilar a los estudiantes mediante variables socioeconómicas, demográficas y conductuales para poder predecir el desempeño académico hasta el fin de año. Además se pide analizar la posibilidad de abstraer en categorias latentes un cojunto de variables ambientales del alumno.

### __Solucion__: 

- Para perfilar a los estudiantes se empleara un módelo predictivo en base a las notas registradas durante los semestres. Para ello se planeta realizar regresiones lineales múltiples por separado y considerando todas las variables presentes (módelos saturados), para luego depurar el modelo en base a la incidencia de cada parametro y el p-value obtenido para cada regresor.

- Para evaluar el ajuste de los modelos, se interpretaran los valores de los estadísticos de bondad de ajuste $R-Cuadrado$ y $Prueba-F$. Además de hacer las pruebas respectivas de eficiencia para los distintos modelos que se generen durante la depuración.


- Por otra parte, en base a los modelos conseguidos se evaluará la incidencia de las categorías latentes pedidas, y se evaluará para los distintos modelos evaluades en la sección anterior.

También se ralizaran las respectivas recodificaciones acorde a las nomenclaturas indicadas y binarozación de las variables dicotómicas.

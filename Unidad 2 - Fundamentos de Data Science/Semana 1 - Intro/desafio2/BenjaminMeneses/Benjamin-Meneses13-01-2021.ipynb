### Desaf칤o 5
<ul>
    <li>Utilizando la misma base <code>flights.csv</code> hacer un loop y clasificar los meses con unacantidad de pasajeros menor a la media</li>
    <li>Para ello, generen un nuevo objeto que represente la media de <code>passengers</code>.</li>
    <li>Generen una columna en la base de datos que se llame <code>underperforming</code> y as칤gnele 0.</li>
    <li>Ejecuten un loop que recorra cada observaci칩n de <code>passengers</code>, donde si la
observaci칩n es menor a la media de <code>passengers</code> se le asigne a <code>underperforming</code>
un 1.</li>
    <li>Para asignar unos en la columna <code>underperforming</code> ya creada, para ello utilice la
funci칩n <code>at</code> de pandas.
</li>
</ul>

```python
import pandas as pd
# Importar el dataframe
df = pd.read_csv('./flights.csv')
# Guardar la media en una variable
mean = df['passengers'].mean()
# Crear la columna en el dataframe y asignarle 0
df['underperforming'] = 0
# Recorrer el dataframe
for idx, row in df.iterrows():
    # Marcar aquellos registros que presentan un valor menor a la media
    if row['passengers'] < mean:
        df.at[idx,'underperforming'] = 1
# Mostrar la frecuencia de 'underperforming'
df['underperforming'].value_counts()
```

### Desafio 6

Hacer un loop que clasifique los meses donde la cantidad de pasajeros se escapa de la
tendencia.
<li>Para ello, genere dos objetos que guarden la media general y la desviaci칩n est치ndar
general de passengers.</li>
<li>Genere una nueva columna en la tabla de datos que se llame outlier y asignele 0.</li>
<li>Ejecuten un loop que recorra cada observaci칩n de passengers, donde si la
observaci칩n se escapa de la tendencia sea 1, de lo contrario 0.</li>
<li>
    Para clasificar los casos que se escapen la tendencia, la observaci칩n debe satisfacer una de las siguientes condiciones:
    <ul>
        <li>La observaci칩n debe ser menor a la media menos la desviaci칩n est치ndar, 洧롇.</li>
        <li>La observaci칩n debe ser mayor a la media m치s la desviaci칩n est치ndar.</li>
        <li>쮺u치ntas observaciones se pueden clasificar como casos extremos?</li>
    </ul>
</li>

```python
stdDev = df['passengers'].std()
mean = df['passengers'].mean()

df['outlier'] = 0
for idx, row in df.iterrows():
    if (row['passengers'] < (mean - stdDev)) or (row['passengers'] > (mean + stdDev)):
        df.at[idx, 'outlier'] = 1

df['outlier'].value_counts()
```
